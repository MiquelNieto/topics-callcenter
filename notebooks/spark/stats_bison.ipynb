Quitamos las Stop Words: 

```python
raw_list_word = raw_bison.map( lambda document: document.strip().lower()) \
                .map( lambda document: re.split(" ", document))

raw_list_word.take(1)
```

```python
raw_no_stop = raw_bison.map( lambda document: document.strip().lower()) \
	.map( lambda document: re.split(" ", document)) \
    .map( lambda word: [x for x in word if x not in StopWords])
raw_no_stop.take(1)
```

```python
list_words = raw_no_stop.reduce(lambda a,b: list(set(a+b)))
```
