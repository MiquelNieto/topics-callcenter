<div style="background-color: #eeffee; border-color: #bbFFbb; border-left: 5px solid #88FF88; padding: 0.5em;">

Cargamos el dataset y vemos la estructura del mismo.
</div>

```python
domo_dataset = sqlContext.read.parquet("dataset/domo_dataset.parquet")
domo_dataset
```

<div style="background-color: #eeffee; border-color: #bbFFbb; border-left: 5px solid #88FF88; padding: 0.5em;">

En principio, nos quedamos unicamente con las llamadas del sistema "Verint" en crudo, aunque también existen sanitizadas (eliminando nombres y datos sensibles). Y mostramos el número total de llamadas con datos.
</div>

```python
raw_verint = domo_dataset.select("raw_verint").rdd.filter(lambda x: x["raw_verint"] is not None)\
    .map(lambda x: " ".join(map(lambda y: " ".join(y), x)))\
    .repartition(17)

raw_verint.cache()
##Llamadas totales buenas
print('Llamadas disponibles : {:,}'.format( raw_verint.count()))  
```

<div style="background-color: #eeffee; border-color: #bbFFbb; border-left: 5px solid #88FF88; padding: 0.5em;">

Número de palabras medio por llamada: 
</div>

```python
import locale
locale.setlocale(locale.LC_ALL, 'es_ES')
avg_words_per_call = raw_verint.map(lambda x: len(x)).reduce(lambda a,b: a+b)/raw_verint.count()
print('Palabras por llamadas: {:,.0f}'.format(avg_words_per_call))  
```

<div style="background-color: #eeffee; border-color: #bbFFbb; border-left: 5px solid #88FF88; padding: 0.5em;">

Pasamos cada llamada(documento) a una lista de palabras.
</div>

```python
raw_list_word = raw_verint.map( lambda document: document.strip().lower()) \
                .map( lambda document: re.split(" ", document))

raw_list_word.take(1)
```

<div style="background-color: #eeffee; border-color: #bbFFbb; border-left: 5px solid #88FF88; padding: 0.5em;">

Eliminamos las Stopwords en español:
</div>

```python
StopWords = stopwords.words("spanish")

raw_no_stop = raw_verint.map( lambda document: document.strip().lower()) \
	.map( lambda document: re.split(" ", document)) \
    .map( lambda word: [x for x in word if x not in StopWords])
raw_no_stop.take(1)
```

<div style="background-color: #eeffee; border-color: #bbFFbb; border-left: 5px solid #88FF88; padding: 0.5em;">

Volvemos a unir las palabras de la lista para mostrar una nube de palabras más frecuentes
</div>

```python
list_words = raw_no_stop.reduce(lambda a,b: list(set(a+b)))

wordcloud = WordCloud().generate(" ".join(list_words))

# Display the generated image:
plt.figure(figsize = (20,20))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

<div style="background-color: #eeffee; border-color: #bbFFbb; border-left: 5px solid #88FF88; padding: 0.5em;">

<div style="background-color: #eeffee; border-color: #bbFFbb; border-left: 5px solid #88FF88; padding: 0.5em;">

A continuación nos quedamos solo con la raiz de las palabras que sean verbos o nombres.
</div>

```python
def get_stem_of_candidates(x):
    good = [u'VLinf', u'NC']
    candidates =  list(filter(lambda word: word in tags_dict.value and tags_dict.value[word][0] in good  ,x))
    stem = list(map(lambda word: tags_dict.value[word][1], candidates))
    return stem
    
stemmed_candidates = raw_no_stop.map(get_stem_of_candidates)    
stemmed_candidates.cache()
stemmed_candidates.count()
```

```python
list_stemmed_words = stemmed_candidates.reduce(lambda a,b: list(set(a+b)))

wordcloud = WordCloud().generate(" ".join(list_stemmed_words))

# Display the generated image:
plt.figure(figsize = (20,20))

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
